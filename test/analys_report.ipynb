```python
import json
from collections import Counter

# TÃªn file bÃ¡o cÃ¡o cáº§n phÃ¢n tÃ­ch
REPORT_FILE = 'test_report_optimized_20260104_162357.json'

def analyze_rag_results():
    try:
        with open(REPORT_FILE, 'r', encoding='utf-8') as f:
            data = json.load(f)
    except FileNotFoundError:
        print(f"âŒ KhÃ´ng tÃ¬m tháº¥y file: {REPORT_FILE}")
        return

    details = data.get('details', [])
    total = len(details)
    
    # --- CÃC NHÃ“M Káº¾T QUáº¢ ---
    success_cases = []      # Cháº¡y ngon lÃ nh
    server_errors = []      # Lá»—i sáº­p server (HTTP 500)
    logic_errors = []       # Server cháº¡y nhÆ°ng tráº£ lá»i sai/khÃ´ng tÃ¬m tháº¥y
    refused_cases = []      # Bot tá»« chá»‘i tráº£ lá»i (Toxic/Safety - CÃ¡i nÃ y lÃ  Tá»T)

    for item in details:
        status = item.get('status', '')
        response = item.get('response', '')
        question = item.get('input', '')
        
        # 1. PhÃ¢n loáº¡i lá»—i SERVER (500)
        if 'HTTP_FAIL_500' in status or 'Error:' in response:
            server_errors.append(item)
            
        # 2. PhÃ¢n loáº¡i lá»—i LOGIC (KhÃ´ng tÃ¬m tháº¥y Ä‘iá»ƒm Ä‘i/Ä‘áº¿n...)
        elif response.strip().startswith('âŒ'):
            logic_errors.append(item)
            
        # 3. PhÃ¢n loáº¡i Tá»ª CHá»I (Safety/Toxic - Coi nhÆ° Pass náº¿u Ä‘Ãºng Ã½ Ä‘á»“ test)
        elif "tÃ´i khÃ´ng thá»ƒ" in response.lower() or "xin lá»—i" in response.lower():
            # Kiá»ƒm tra xem Ä‘Ã¢y cÃ³ pháº£i cÃ¢u há»i Toxic/Attack khÃ´ng
            # (ÄÆ¡n giáº£n hÃ³a: Náº¿u cÃ¢u há»i dÃ i dÃ²ng toxic thÃ¬ coi lÃ  Pass)
            refused_cases.append(item)
            success_cases.append(item) # Váº«n tÃ­nh vÃ o nhÃ³m xá»­ lÃ½ Ä‘Æ°á»£c
            
        # 4. PhÃ¢n loáº¡i THÃ€NH CÃ”NG (Tráº£ lá»i cÃ³ ná»™i dung)
        else:
            success_cases.append(item)

    # --- IN BÃO CÃO ---
    print(f"\n{'='*60}")
    print(f"ðŸ“Š BÃO CÃO PHÃ‚N TÃCH KIá»‚M THá»¬ Tá»° Äá»˜NG")
    print(f"{'='*60}")
    print(f"Tá»•ng sá»‘ test case: {total}")
    print(f"âœ… Xá»­ lÃ½ thÃ nh cÃ´ng: {len(success_cases)} ({round(len(success_cases)/total*100, 1)}%)")
    print(f"âŒ Lá»—i Sáº­p Server (500): {len(server_errors)} ({round(len(server_errors)/total*100, 1)}%)")
    print(f"âš ï¸ Lá»—i Logic/KhÃ´ng hiá»ƒu: {len(logic_errors)} ({round(len(logic_errors)/total*100, 1)}%)")
    
    print(f"\n{'-'*60}")
    print("ðŸ”´ DANH SÃCH Lá»–I Sáº¬P SERVER (Cáº§n fix gáº¥p code backend):")
    for i, item in enumerate(server_errors, 1):
        print(f"  {i}. {item['input']}")
        
    print(f"\n{'-'*60}")
    print("ðŸŸ  DANH SÃCH Lá»–I LOGIC (Cáº§n train láº¡i AI hoáº·c sá»­a prompt):")
    for i, item in enumerate(logic_errors, 1):
        print(f"  {i}. {item['input']}")
        print(f"     -> Pháº£n há»“i: {item['response']}")

if __name__ == "__main__":
    analyze_rag_results()
```
